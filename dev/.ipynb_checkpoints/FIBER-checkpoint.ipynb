```python
def detect_peaks(time,signal,window):
    # calculate zscores
    signal = (signal - signal.mean())/signal.std()
    # find indexes for n second windows
    time_points = np.arange(time[0],time[-1],window)
    if time_points[-1] != time[-1]: time_points = np.concatenate((time_points,[time[-1]]))
    indexes = [np.where(abs(time-i) == abs(time-i).min())[0][0] for i in time_points]
    signal_bins = [signal[indexes[i-1]:indexes[i]] for i in range(1,len(indexes))]
    time_bins = [time[indexes[i-1]:indexes[i]] for i in range(1,len(indexes))]
    bin_mad = []
```
